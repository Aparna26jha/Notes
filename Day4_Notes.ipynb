## 🧠 **1. SENTENCE TRANSFORMERS**

### What are Sentence Transformers?
- **Pre-trained models** that convert text into dense numerical vectors (embeddings)
- Built on top of transformer architectures (like BERT, RoBERTa)
- Specifically designed to create **semantically meaningful sentence embeddings**

### Key Model: `all-MiniLM-L6-v2`


```python

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```



**Model Specifications:**
- **Max sequence length**: 256 tokens
- **Embedding dimension**: 384 dimensions
- **Provider**: sentence-transformers library
- **Purpose**: General-purpose semantic similarity

### How It Works:
1. **Tokenization**: Text → tokens
2. **Transformer layers**: Process contextual relationships
3. **Pooling**: Create fixed-size representation
4. **Output**: 384-dimensional vector capturing semantic meaning

## 🔢 **2. EMBEDDINGS**

### Definition
Embeddings are **dense numerical vectors** that represent text in a high-dimensional space where semantically similar content is positioned closer together.

### Types of Embeddings in our Code:

#### A) **TF-IDF Embeddings** (Traditional)


```python
vectorizer = TfidfVectorizer(
    max_features=1000,
    stop_words='english', 
    ngram_range=(1, 2)  # 1-word and 2-word phrases
)
```



**Features:**
- **Sparse vectors** (many zeros)
- Based on term frequency and inverse document frequency
- Good for keyword matching

## 🔍 **4. RAG (Retrieval-Augmented Generation)**

### What is RAG?
RAG combines **information retrieval** with **text generation** to provide accurate, contextual responses based on a knowledge base.

### RAG Pipeline:

#### Step 1: **Knowledge Base Preparation**


```python
# From your medical example
medical_docs = [
    "Drug: aspirin for cardiovascular protection",
    "Drug: metformin for diabetes glucose control",
    "Condition: hypertension high blood pressure"
]
```



#### Step 2: **Embedding Creation**


```python
# Convert documents to vectors
document_vectors = vectorizer.fit_transform(medical_docs)
```



#### Step 3: **Query Processing**


```python
# User query
query = "heart medication"

# Convert query to same vector space
query_vector = vectorizer.transform([query])
```



#### Step 4: **Similarity Search**


```python
# Find most relevant documents
similarities = cosine_similarity(query_vector, document_vectors)
best_match_idx = np.argmax(similarities)
```



#### Step 5: **Context Retrieval**


```python
# Get relevant documents for LLM context
relevant_docs = [medical_docs[best_match_idx]]
```



### RAG Advantages:

1. **Semantic Understanding**:
   - Query: "heart medication" 
   - Matches: "cardiovascular protection" (synonym understanding)

2. **Dynamic Knowledge**:
   - No model retraining needed
   - Update knowledge base directly

3. **Source Attribution**:
   - Responses traceable to specific documents
   - Confidence scores available

**Limitations:**
- Misses synonyms ("cardiovascular" ≠ "heart")
- No context understanding
- Exact word matches only
